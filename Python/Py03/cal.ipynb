```python
import math
```

```python
def calculator():
  is_continue = 'y'
  while is_continue != 'q':
    user_input = input("수식을 입력하세요: ").strip()
    target_op = None
    for operator in ['**', '+', '-', '/', '*']:
        if operator in user_input:
           target_op = operator
           index = user_input.find(operator)
           break
    if target_op == None:
      print("지원하지 않는 연산자입니다.")
      continue
    try:
      num1 = int(user_input[:index])
      num2 = int(user_input[index + len(target_op):])
      if target_op == '**':
        result = num1**num2
      elif target_op == '+':
        result = num1 + num2
      elif target_op == '-':
        result = num1 - num2
      elif target_op == '/':
        result = num1 / num2
      elif target_op == '*':
        result = num1 * num2
      else:
        raise ValueError
    except ValueError:
      print("잘못된 입력입니다. 정수를 입력해주세요.")
    except ZeroDivisionError:
      print("0으로 나눌 수 없습니다.")
    else:
      print(f"계산결과: {result}")
    finally:
      is_continue = input("계산을 종료 하시겠습니까? q를 입력하세요.").strip()
      if is_continue == 'q':
        print("계산을 종료합니다.")

```

```python
calculator()
```
